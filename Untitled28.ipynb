```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import os
```

```python
img = cv2.imread('/content/veg-bloggers-lead.webp',cv2.IMREAD_GRAYSCALE)
m,n = img.shape
crow, ccol = m//2, n//2
print(m,n)
```

```python
dft = np.fft.fft2(img)
dft_shift = np.fft.fftshift(dft)
```

```python
def ideal_lpf(size, cutoff):
    rows, cols = size
    mask = np.zeros((rows, cols), np.uint8)
    for i in range(rows):
        for j in range(cols):
            if np.sqrt((i - crow)**2 + (j - ccol)**2) < cutoff:
                mask[i, j] = 1
    return mask
```

```python
plt.imshow(ideal_lpf((m,n), 40))
```

```python
def ideal_hpf(size, cutoff):
    return 1 - ideal_lpf(size, cutoff)

plt.imshow(ideal_hpf((m,n), 40))
```

```python
def butterworth_lpf(size, cutoff, order):
    rows, cols = size
    mask = np.zeros((rows, cols))
    for i in range(rows):
        for j in range(cols):
            d = np.sqrt((i - crow)**2 + (j - ccol)**2)
            mask[i, j] = 1 / (1 + (d / cutoff)**(2 * order))
    return mask
plt.imshow(butterworth_lpf((m,n), 40,2))
```

```python
def butterworth_hpf(size, cutoff, order):
    return 1 - butterworth_lpf(size, cutoff, order)
plt.imshow(butterworth_hpf((m,n), 40,2))
```

```python
def gaussian_lpf(size, cutoff):
    rows, cols = size
    mask = np.zeros((rows, cols))
    for i in range(rows):
        for j in range(cols):
            d = np.sqrt((i - crow)**2 + (j - ccol)**2)
            mask[i, j] = np.exp(-(d**2) / (2 * cutoff**2))
    return mask
plt.imshow(gaussian_lpf((m,n), 40))
```

```python
def gaussian_hpf(size, cutoff):
    return 1 - gaussian_lpf(size, cutoff)
plt.imshow(gaussian_hpf((m,n), 40))
```

```python
def apply_filter(mask):
    filtered = dft_shift * mask
    ishift = np.fft.ifftshift(filtered)
    img_back = np.fft.ifft2(ishift)
    img_back = np.abs(img_back)
    return img_back
```

```python
cutoff = 30
order = 2
ideal_LPF = ideal_lpf(img.shape, cutoff)
ideal_HPF = ideal_hpf(img.shape, cutoff)

butter_LPF = butterworth_lpf(img.shape, cutoff, order)
butter_HPF = butterworth_hpf(img.shape, cutoff, order)

gauss_LPF = gaussian_lpf(img.shape, cutoff)
gauss_HPF = gaussian_hpf(img.shape, cutoff)


ideal_LPF_img = apply_filter(ideal_LPF)
ideal_HPF_img = apply_filter(ideal_HPF)

butter_LPF_img = apply_filter(butter_LPF)
butter_HPF_img = apply_filter(butter_HPF)

gauss_LPF_img = apply_filter(gauss_LPF)
gauss_HPF_img = apply_filter(gauss_HPF)
```

```python
titles = [
    "Original",
    "Ideal LPF", "Ideal HPF",
    "Butterworth LPF", "Butterworth HPF",
    "Gaussian LPF", "Gaussian HPF"
]

images = [
    img,
    ideal_LPF_img, ideal_HPF_img,
    butter_LPF_img, butter_HPF_img,
    gauss_LPF_img, gauss_HPF_img
]

plt.figure(figsize=(12, 10))
for i in range(7):
    plt.subplot(3, 3, i+1)
    plt.imshow(images[i], cmap="gray")
    plt.title(titles[i])
    plt.axis("off")
```
